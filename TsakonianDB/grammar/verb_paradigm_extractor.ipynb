# A paradigms

Verbs ending in `-ου`

```python
# A1
ending = 'νου'
negative_ending = 'ούνου'

paradigm = 'Α1'
add = True
mask = (df['tsakonian'].str.endswith(ending) & ~df['tsakonian'].str.endswith(negative_ending))

if add:
    df.loc[mask, 'paradigm'] = paradigm
display(df[mask])
```
